# Task 1: Draw photo using OpenCV and Numpy

```python
import cv2
import numpy as np
```

```python
# Create 3 D array with zeros
photo = np.zeros((550, 500, 3))
```

```python
# Draw face
cv2.circle(photo, center=(250,130), radius=100, color=(240,255,100), thickness= -1)
```

```python
# Draw eyes
cv2.circle(photo, center=(215,120), radius=15, color=(0, 0, 0), thickness= -1)
cv2.circle(photo, center=(275,120), radius=15, color=(0, 0, 0), thickness= -1)
```

```python
# Draw mouth
cv2.circle(photo, center=(250,180), radius=47, color=(0, 0, 0), thickness= -1)
```

```python
# Draw Body Part
cv2.circle(photo,(245,300),150,[240,255,100],-1)
```

```python
# Draw legs
cv2.line(photo, (250, 390), (150, 500), (240,255,100), 30)
cv2.line(photo, (250, 390), (350, 500), (240,255,100), 30)
```

```python
# Put the text 
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(photo,'Hello',(140,300), font, 3,(0, 0, 0),6,cv2.LINE_AA)
cv2.putText(photo,'OpenCV',(140,380), font, 2,(0, 0, 0),6,cv2.LINE_AA)
```

```python
# Display image
cv2.imshow("Picture", photo)
cv2.waitKey()
cv2.destroyAllWindows()
```
